```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np

import costs
from plotting.costs import plot_costs
```

```python
num_features=4
```

```python
cc = costs.ConstantCosts(
    feature_vector_size=num_features,
    cost_values=list(range(1, num_features+1)),
)
```

```python
plot_costs(cc, 100, 'Constant costs')
```

```python
bc = costs.BernoulliCosts(
    feature_vector_size=num_features,
    probabilities=[0.85]*num_features,
    cost_values=list(range(1, num_features+1)),
    random_seed=42
)
```

```python
plot_costs(bc, 100, 'Bernoulli costs')
```

```python
nbc = costs.NonstationaryBernoulliCosts(
    feature_vector_size=num_features,
    change_points=[30, 60],
    interval_probabilities=[
        [0.85]*num_features,
        [0.95]*num_features,
        [0.35]*num_features,
    ],
    interval_cost_values=[
        [1,2,3,4],
        [2,4,6,8],
        [1,7,2,3]
    ],
    random_seed=42
)
```

```python
plot_costs(nbc, 100, 'Non-stationary Bernoulli')
```

```python
gc = costs.GaussianCosts(
    feature_vector_size=num_features,
    means=list(range(1, num_features+1)),
    stds=[0.25]*num_features,
    random_seed=42,
)
```

```python
plot_costs(gc, 100)
```

```python
ngc = costs.NonstationaryGaussianCosts(
    feature_vector_size=num_features,
    change_points=[40, 60],
    interval_means=[
        [1,2,3,4],
        [2,4,6,8],
        [1,2,4,8]
    ],
    interval_stds=[
        [0.25]*num_features,
        [0.25]*num_features,
        [0.05]*(num_features-1)+[1.5],
    ],
    random_seed=42,
)
```

```python
plot_costs(ngc, 100)
```
