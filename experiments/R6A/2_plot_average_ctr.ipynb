```python
%load_ext autoreload
%autoreload 2

from datetime import datetime
import numpy as np
import pickle
```

```python
with open('total_click_through_rates.pickle', 'rb') as f:
    article_ctrs = pickle.load(f)
```

```python
SUBSAMPLE_PERCENTAGE = 1

small_ctr = {k:v for k,v in article_ctrs.items() if np.random.binomial(n=1, p=SUBSAMPLE_PERCENTAGE)}
```

```python
import plotly.graph_objects as go

fig = go.Figure()

for art_id in sorted(small_ctr):
    ctr = small_ctr[art_id][0]
    tss = small_ctr[art_id][1]
    times = [datetime.fromtimestamp(ts) for ts in tss]
    fig.add_scatter(x=times, y =ctr, mode='lines', name=int(art_id))
    
    
fig
```

```python
fig.write_html('average_ctr_plot.html')
```
